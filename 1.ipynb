```python
#GCD

#taking input 

def gcd(temp1,temp2):
    while (temp2/temp1)!=0:
        rem=temp1%temp2
        temp1=temp2
        temp2=rem
    return temp1

n1,n2,n3=map(int,input("Enter the three numbers : ").split())
ans1=gcd(n1,n2)
ans2=gcd(ans1,n3)
print(ans1)
    
```

```python
#creating finction bubble sort

def bubble_sort(list2):
    for i in range(len(list2)-1,0,-1):
        for j in range(i):
            if list2[j]>list2[j+1]:
                temp=list2[j]
                list2[j]=list2[j+1]
                list2[j+1]=temp
    return list2

list1=[2,4,1,7,6]
print(len(list1))
bubble_sort(list1)
print(list1)
```
